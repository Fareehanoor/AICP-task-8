```python
from datetime import datetime, timedelta

class Boat:
    def __init__(self, boat_number):
        self.boat_number = boat_number
        self.money_taken = 0
        self.total_hours = 0
        self.return_time = None

def calculate_money_for_one_boat(boat):
    current_time = datetime.now().replace(hour=10, minute=0, second=0, microsecond=0)
    closing_time = datetime.now().replace(hour=17, minute=0, second=0, microsecond=0)

    if current_time < datetime.now().replace(hour=10, minute=0, second=0, microsecond=0):
        print("Boats cannot be hired before 10:00.")
        return

    if current_time > datetime.now().replace(hour=17, minute=0, second=0, microsecond=0):
        print("Boats cannot be returned after 17:00.")
        return

    duration = float(input("Enter the duration (0.5 for half an hour, 1 for one hour): "))

    cost_per_hour = 20
    cost_per_half_hour = 12

    if duration == 1:
        cost = cost_per_hour
    elif duration == 0.5:
        cost = cost_per_half_hour
    else:
        print("Invalid duration. Please choose 0.5 or 1.")
        return

    boat.money_taken += cost
    boat.total_hours += duration
    return_time = current_time + timedelta(hours=duration)
    boat.return_time = return_time

    print(f"Boat {boat.boat_number} hired for {duration} hours. Money taken: ${cost:.2f}")

def find_next_available_boat(boats):
    current_time = datetime.now().replace(hour=10, minute=0, second=0, microsecond=0)

    available_boats = [boat for boat in boats if boat.return_time is None or boat.return_time < current_time]

    if not available_boats:
        next_available_time = min([boat.return_time for boat in boats])
        print(f"No boats available. Next available time: {next_available_time}")
    else:
        available_boat_numbers = [boat.boat_number for boat in available_boats]
        print(f"Available boats: {available_boat_numbers}")

def calculate_money_for_all_boats(boats):
    total_money = sum([boat.money_taken for boat in boats])
    total_hours = sum([boat.total_hours for boat in boats])
    unused_boats = [boat.boat_number for boat in boats if boat.return_time is None]

    max_used_boat = max(boats, key=lambda x: x.total_hours, default=None)

    print(f"\nTotal money taken for all boats: ${total_money:.2f}")
    print(f"Total hours boats were hired: {total_hours} hours")
    print(f"Boats not used today: {unused_boats}")
    if max_used_boat:
        print(f"Boat {max_used_boat.boat_number} was used the most today.")

def main():
    boats = [Boat(i) for i in range(1, 11)]

    # TASK 1 - calculate the money taken in a day for one boat.
    boat_number = int(input("Enter the boat number (1 to 10): "))
    if 1 <= boat_number <= 10:
        calculate_money_for_one_boat(boats[boat_number - 1])

    # TASK 2 - find the next boat available.
    find_next_available_boat(boats)

    # TASK 3 - calculate the money taken for all the boats at the end of the day.
    calculate_money_for_all_boats(boats)

if __name__ == "__main__":
    main()
```
